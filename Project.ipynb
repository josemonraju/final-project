```python
#function 2 : take a list and count the words in it to make a dictionary
# Code from:  https://stackoverflow.com/questions/21852066/counting-word-frequency-and-making-a-dictionary-from-it 

def list_dictionary(listvar):
    dict={}
    
    for word in listvar:#loop
        if word in dict:# checks if word is already in the dictionary
            dict[word]+=1
        else:
            dict[word]=1
    
    return dict
  
```

```python
# function 3 v2 - Source: Suresh; JR added some error handling 

import re, datetime

def stringtoyear(title):
    try:
        match = re.search('\d{4}', title)
        extract_date = datetime.datetime.strptime(match.group(), '%Y').date()
        return int(extract_date.year)
    
    except AttributeError as Error: # added error handling for when there is no year mentioned.
        #match = re.search('\d{4}', '1111')
        #extract_date = datetime.datetime.strptime(match.group(), '%Y').date()
        return int('0000') 


```

```python
#https://thispointer.com/pandas-loop-or-iterate-over-all-or-certain-columns-of-a-dataframe/
# Iterate over two given columns only from the dataframe

desc_series=wine[wine['variety']=='Pinot Noir'].head(2000) # artificially limited the amount of data to meet data limits.
combined_list=[]


for column in desc_series['description']:
    # Select column contents by column name using [] operator
    #columnSeriesObj = column
    #x=columnSeriesObj.values #some sort of arrary
    
    for item in column: # for each list item in the column cells (array)
        
        combined_list.append(item)

print (combined_list)
```

```python
#conda update -n base -c defaults conda
```

```python
wine_flavors_count=list_dictionary(combined_list)# creates a new dictionary with the list of descriptions count functions

sorted(wine_flavors_count.items(), key=lambda item: item[1],reverse=True)# sorted function 
#https://stackoverflow.com/questions/4088265/sorted-word-frequency-count-using-python #+ von

```

```python
# Turns out it's easy to make a wordcloud once you have a dictionary. Just install the wordcloud libary and import your frequency list.
```

Word cloud of Pinot Noir descriptions

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

wordcloud.generate_from_frequencies(wine_flavors_count)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.show()
```

```python
# Note the wordcloud tutorial happened to use the same data set as i did from Kaggle. 
# This is a super easy option. No making of a dictionary needed. Just feed it the original strings.
#https://www.datacamp.com/community/tutorials/wordcloud-python
```

```python
#https://www.datacamp.com/community/tutorials/wordcloud-python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
#Goal: Make a visual word cloud that shows the words used to describe a certain variety of wine. 

wine=pd.read_csv('wine.csv', header=0)
# reimporting the wine dataframe since the tutorial used doesn't require the descriptions to be modified. They can be kept as strings.

text = " ".join(review for review in wine[wine['variety']=='Pinot Noir'].description)

# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update({"drink", "now", "wine", "flavor", "Pinot", " Noir", "the"})

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.show()
```
